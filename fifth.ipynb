```python
# 뉴스 검색 전체코드
import requests
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/news/mainnews.naver'
# url = 'https://finance.naver.com/news/mainnews.naver?date'+date  # 날짜를 지정하여 검색하는경우

response = requests.get(url)
print(response.status_code)

html = response.text
soup = BeautifulSoup(html, 'html.parser')

# 기사제목, 기사요약, 신문사, 작성날짜, 링크
articles = soup.select(".block1")

data = []
for article in articles:
    title = article.select_one('.articleSubject > a').text
    summary = article.select_one('.articleSummary').contents[0].text.strip()
    press = article.select_one('.press').text.strip()
    wdate = article.select_one('.wdate').text
    link = article.select_one('.articleSubject > a').attrs['href']
    article_id = link.split('=')[1].split('&')[0]
    office_id=link.split('office_id=')[1].split('&')[0]
    link = f"https://n.news.naver.com/mnews/article/{office_id}/{article_id}"

    # print({"title":title, "summary": summary, "press": press, "wdate": wdate, "link": link})
    data.append({"title":title, "summary": summary, "press": press, "wdate": wdate, "link": link})
```

```python
import requests

url = 'https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade'
params = {
    'LAWD_CD': '41595',
    'DEAL_YMD': '202501',
    'serviceKey': 'your_apikey'
}

response = requests.get(url,params=params)
response
```

```python
import xmltodict
response_data = xmltodict.parse(response.content)
response_data
```

```python
items = response_data['response']['body']['items']['item']

# 아파트명, 법정동, 거래금액, 전용면적, 층, 건축년도
item = items[0]
{'아파트명': item['aptNm'], '법정동': item['umdNm'], '거래금액': item['dealAmount'], '전용면적': item['excluUseAr'], '층': item['floor'], '건축년도':item['buildYear'] }
```

```python
apt_data = []
for item in items:
    apt_data.append({'아파트명': item['aptNm'], '법정동': item['umdNm'], '거래금액': item['dealAmount'], '전용면적': item['excluUseAr'], '층': item['floor'], '건축년도':item['buildYear'] })
apt_df = pd.DataFrame(apt_data)
apt_df.to_csv('data/동탄아파트실거래가202501.csv')
```
