```python
# 복습 - 1부터 10까지의 자연수 중에서 홀수와 홀수들의 합
# 아래 출력예시
# 1, 3, 5, 7, 9
# 1~10까지의 홀수의 합: 25

hap = 0
for i in range(1, 11):
    if i % 2 == 1:
        print(i, end=', ')
        hap += i
# print()
print(f'\n1~10까지의 홀수의 합: {hap}')
```

```python
hap = 0
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i, end=', ')
    hap += i

print(f'\n1~10까지의 홀수의 합: {hap}')
```

```python
# 20보다 작은 정수 중에서 3의 배수와 5의 배수를 제외하고 출력
x = 0

while True:
    x += 1
    if x > 20:
        break
    if (x % 3 == 0) or (x % 5 == 0):
        continue
    print(x, end=' ')
    
# 1 2 4  ....
```

```python
# 컴퓨터와 대결하는 가위, 바위, 보 게임
# '그만'을 입력할때까지 계속
# '이겼습니다' '졌습니다' '비겼습니다' (전적: 승 패)
import random

sel = ['가위', '바위', '보']

while True:
    com = random.randint(0, 2)
    print(sel[com])

    user = input('"가위", "바위", "보" 중 하나를 입력하세요(종료하려면 "그만" 입력): ')
    print(f"{sel[com]} vs {user}")

    if user == '그만':
        print('게임을 종료')
        break

    if user not in sel:
        print('제대로 입력하세요!!')
        continue

    if user == sel[com]:
        print('비겼어요')
    elif (user == '가위' and sel[com] == '보') or (user == '바위' and sel[com] == '가위') or (user == '보' and sel[com] == '바위'):
        print('이겼어요')
    else:
        print('졌어요')
```

```python
# 로또번호 추첨기
# 1부터 45까지 6개의 중복되지 않는 번호를 추첨 append 
# 오름차순으로 나열해서 출력 sort
import random

numbers = []

while len(numbers) < 6:
    number = random.randint(1, 45)
    if number not in numbers:
        numbers.append(number)
numbers.sort()

print('이번 주 로또 번호: ', numbers)
```

```python
addr = {
    '박지민': '010-1111-1111', '전정국': '010-2222-2222',
    '민윤기': '010-3333-3333', '정호석': '010-4444-4444',
    '김석진': '010-5555-5555', '김태형': '010-6666-6666'
}

while True:
    menu = int(input("메뉴선택(1: 추가, 2: 검색, 0: 종료)"))

    if menu == 0:
        print('프로그램 종료')
        break
    elif menu == 2:
        print('검색 가능한 친구들: ', list(addr.keys()))
        name = input('친구 이름 입력: ')
        if name in addr:
            print(f"{name}의 전화번호: {addr[name]}")
        else:
            print('등록되지 않은 친구입니다.')
    elif menu == 1:
        addr_key = input('추가할 친구 이름 입력: ')
        addr_value = input('전화번호 입력: ')
        addr[addr_key] = addr_value
    else:
        print('메뉴 선택 오류')
```
