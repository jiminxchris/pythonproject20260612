```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
plt.figure(figsize=(5, 3)) # 그래프 크기 조절

plt.plot([1, 2, 3],[4,5,6],label = 'graph1') # 그래프의 종류와 좌표 찍기
plt.plot([4, 5, 6], [1, 2, 3], label = 'graph2')

plt.legend()
plt.title('선 그래프') # 제목달기
plt.xlabel('x축') # x축 이름 달기
plt.ylabel('y축') # y축 이름 달기
plt.grid()
plt.show() # 그래프 그리기
```

```python
# 막대그래프, 히스토그램, 산점도 
bar, hist, scatter 
```

```python
plt.bar(['박지민', '전정국', '민윤기'], [10, 20, 30])

plt.xlabel('이름')
plt.ylabel('점수')
plt.title('막대 그래프')
plt.show()
```

```python
plt.hist([10, 20, 20, 30, 30, 30], bins=3)

plt.xlabel('점수')
plt.ylabel('빈도')
plt.title('히스토그램')
plt.show()
```

```python
plt.scatter([1, 2, 3], [4, 5, 6])

plt.xlabel('x축')       
plt.ylabel('y축')
plt.title('산점도') 
plt.show()
```

```python
sns.barplot(x=['박지민', '전정국', '민윤기'], y=[10, 20, 30])
plt.title('Seaborn 막대 그래프')
plt.show()   
```

```python
house = pd.read_csv('data/동탄아파트실거래가202501.csv', index_col=0)
```

```python
house
```

```python
# house 데이터의 동별 거래가 평균 구하기
house['거래금액'] = house['거래금액'].str.replace(',', '').astype(float)
```

```python
# house 데이터의 동별 거래가 평균 시각화
house.groupby('법정동')['거래금액'].mean().plot(kind='bar')
plt.xlabel('법정동')
plt.ylabel('평균 거래가')
plt.title('동별 거래가 평균')
plt.show()
```
