```python
def max_num(a, b, c): # 가장 큰 수를 반환하는 함수
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val
    
```

```python
def min_num(num_list):
    min_val = num_list[0]
    for num in num_list:
        if min_val > num:
            min_val = num
    return min_val
```

```python
def isLeapYear(year):
    # 4로 나누어 떨어지는 해 year % 4 == 0
    # 100으로 나누어 떨어지지 않는 해 year % 100 != 0
    # 400으로 나누어 떨어지는 해  (year % 400 == 0)
    return (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0)
```

```python
isLeapYear(2000)  # True
```

```python
등수 구하기

영어 점수와 수학 점수의 평균 점수를 기준으로 학생들의 등수를 매기려고 합니다. 영어 점수와 수학 점수를 담은 2차원 리스트 score가 주어질 때, 영어 점수와 수학 점수의 평균을 기준으로 매긴 등수를 담은 리스트를 return하도록 함수를 만들어주세요

[[80, 70], [90, 50], [40, 70], [50, 80]] -> [1, 2, 4, 3]
[[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]] -> [4, 4, 6, 2, 2, 1, 7]
```

```python
def rank(scores):
    avg = [sum(s)/len(s) for s in scores]
    avg_sort = sorted(avg, reverse=True)
    result = [avg_sort.index(i) + 1 for i in avg]
    return result # [ 1, 2, 3, 4]
```

```python
scores = [[80, 70], [90, 50], [40, 70], [50, 80]]
rank(scores) # [ 1, 2, 3, 4]
```
