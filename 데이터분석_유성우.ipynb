# 데이터 분석

```python
도시의 위도를 확인 cities
위도를 사용해서 해당 도시에 표시되는 별자리 확인 constellations
별자리를 사용해서 해당 도시에서 볼수 있는 유성우 확인 meteor_showers
유성우를 사용해서 해당 유성우를 볼 수 있는 날짜 확인
날짜를 사용해서 달의 위상값이 가장 작은 최적 날짜 확인 moon_phases

마지막으로 사용자에게 기분좋은 문구 반환

cities -> constellations -> meteor_showers -> moon_phases
      위도정보           별자리 -> 유성우    관측가능기간    가장 어두운날
```
