# 데이터 분석

```python
도시의 위도를 확인 cities
위도를 사용해서 해당 도시에 표시되는 별자리 확인 constellations
별자리를 사용해서 해당 도시에서 볼수 있는 유성우 확인 meteor_showers
유성우를 사용해서 해당 유성우를 볼 수 있는 날짜 확인
날짜를 사용해서 달의 위상값이 가장 작은 최적 날짜 확인 moon_phases

마지막으로 사용자에게 기분좋은 문구 반환

cities -> constellations -> meteor_showers -> moon_phases
      위도정보           별자리 -> 유성우    관측가능기간    가장 어두운날
```

## 문자 안내 출력

```python
# 도시명을 입력히면 위도와 별자리리스트와 메시지를 반환하는 함수

def predict_best_meteor_shower_viewing(city):

    meteor_shower_string = ''

    if city not in cities['city'].values:
        meteor_shower_string = f'아쉽지만 {city}에는 별자리 서비스가 제공되지 않습니다!.'
        return meteor_shower_string

    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]

    constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (latitude >= constellations['latitudeend']), 'constellation'].to_list()

    return latitude, constellations_list, meteor_shower_string
```

## 최적의 유성우 관측일

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기

def predict_best_meteor_shower_viewing(city):
    
    meteor_shower_string = ''

    if city not in cities['city'].values:
        meteor_shower_string = f'아쉽지만 {city}에는 별자리 서비스가 제공되지 않습니다!.'
        return meteor_shower_string

    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]

    constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (latitude >= constellations['latitudeend']), 'constellation'].to_list()

    meteor_shower_string = f'{city} 에서는 아래의 유성우들을 관측할 수 있습니다.\n'

    for i in constellations_list:
        meteor_shower = meteor_showers.loc[meteor_showers['radiant'] == i, 'name'].iloc[0]
        meteor_shower_startdate = meteor_showers.loc[meteor_showers['radiant'] == i, 'startdate'].iloc[0]
        meteor_shower_enddate = meteor_showers.loc[meteor_showers['radiant'] == i, 'enddate'].iloc[0]
        moon_phases_list = moon_phases[(moon_phases['date'] >= meteor_shower_startdate) & (moon_phases['date'] <= meteor_shower_enddate)]
        best_moon_date = moon_phases_list.loc[moon_phases_list['percentage'].idxmin(), 'date']
        best_moon_date_str = best_moon_date.to_pydatetime().strftime('%Y년 %m월 %d일')

        meteor_shower_string += f"\n{meteor_shower}는 {best_moon_date_str}에 가장 잘 볼 수 있습니다."

    return meteor_shower_string
```
